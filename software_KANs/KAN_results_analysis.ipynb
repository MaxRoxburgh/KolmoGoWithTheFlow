# Analysis of software KAN results on different tasks

```python
import os
import json
import matplotlib.pyplot as plt

def extract_json_data(folder_path):
    names = []
    widths = []
    accuracies = []
    grids = []

    # List all files in the folder
    for filename in os.listdir(folder_path):
        if filename.lower().endswith(".json"):
            file_path = os.path.join(folder_path, filename)

            try:
                with open(file_path, "r") as f:
                    data = json.load(f)

                # Extract required fields (if present)
                names.append(data.get("task"))
                widths.append(data.get("width"))
                accuracies.append(data.get("test_acc"))
                grids.append(data.get("grid"))


            except (json.JSONDecodeError, FileNotFoundError) as e:
                print(f"Skipping {filename}: {e}")

    return names, widths, accuracies, grids

names, widths, accuracies, grids = extract_json_data("data")
```

```python
def plot_task_accuracies(task_names, accuracies, indicies, title):
    """
    Plots a bar chart of tasks vs. accuracies.

    Parameters:
    task_names (list of str): The names of the tasks.
    accuracies (list of float): The corresponding accuracies.
    """
    if len(task_names) != len(accuracies):
        raise ValueError("task_names and accuracies must have the same length.")

    names = []
    acc = []
    for i in indicies:
        names.append(task_names[i])
        acc.append(accuracies[i])

    plt.figure(figsize=(8, 5))
    plt.title(title)
    plt.bar(names, acc)
    plt.xticks(rotation=45, ha='right')
    plt.ylim([0.5,1.])
    plt.ylabel("Accuracy")
    plt.tight_layout()
    plt.show()

width_one = [2,1,1]
width_two = [2,2,1]
width_three = [2,3,1]

plt_one_indexs = []
plt_two_indexs = []
plt_three_indexs = []
for index, width in enumerate(widths):
    if width_one == width and grids[index] == 5:
        plt_one_indexs.append(index)
    elif width_two == width and grids[index] == 5:
        plt_two_indexs.append(index)
    elif width_three == width:
        plt_three_indexs.append(index)

plot_task_accuracies(names, accuracies, plt_one_indexs, f"KAN ({width_one}) test accuracies")

plot_task_accuracies(names, accuracies, plt_two_indexs, f"KAN ({width_two}) test accuracies")

plot_task_accuracies(names, accuracies, plt_three_indexs, f"KAN ({width_three}) test accuracies")
```

```python
width_three_g10 = [2,1,1]

plt_w2_g10_indexes = []
for index, width in enumerate(widths):
    if width_three_g10 == width and grids[index] == 10:
        plt_w2_g10_indexes.append(index)

plot_task_accuracies(names, accuracies, plt_w2_g10_indexes, f"KAN ({width_three_g10}) test accuracies - grid 10")
```
